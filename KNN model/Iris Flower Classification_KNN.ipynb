```
import pandas as pd
```

```
data = pd.read_csv("/content/Iris.csv")
data.head()
```

```
data.info()
```

```
### Feature and targets
#x = data[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].values
x = data.iloc[:,1:5].values
y = data.iloc[:,5].values
```

```
type(y)
```

```
### Split the data into training and testing
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.80,random_state=3)
```

```
import numpy as np
np.unique(y,return_counts=True)
```

```
### Build the model
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 10)
```

```
model.fit(xtrain,ytrain)
```

```
ypred = model.predict(xtest)
```

```
ypred
```

```
ytest
```

```
df = pd.DataFrame()
df["Actual"] = ytest
df["Predicted"] = ypred
df
```

```
count = 0
for i in range(len(ytest)):
  if ytest[i] == ypred[i]:
    count = count + 1
```

```
count
```

```
wrong_count = len(ytest) - count
wrong_count
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cm
```

```
from sklearn.metrics import accuracy_score
accu = accuracy_score(ytest,ypred)
accu
```

```
accuracy = []
for k in range(3,101):
  m = KNeighborsClassifier(n_neighbors = k)
  m.fit(xtrain,ytrain)
  yp = m.predict(xtest)
  ac = accuracy_score(ytest,yp)
  accuracy.append(ac)
```

```
accuracy
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(15,6))
plt.plot(range(3,101),accuracy,label="K vs Accuracy Graph"
    ,linestyle="dashed",marker="o",markerfacecolor="red")
plt.show()
```
